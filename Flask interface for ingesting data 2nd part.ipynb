```python
import random
import requests

number_books = 100

titles = ['cat 1', 'cat 2', 'cat 3', 'cat 4', 'cat 5', 'cat 6']
authors = ['Author 7', 'Author 8', 'Author 9', 'Author 10', 'Author 11', 'Author 12']
years_of_publication = list(range(2010, 2024))

for _ in range(number_books):
    title = random.choice(titles)
    author = random.choice(authors)
    year_of_publication = random.choice(years_of_publication)


    print(f"Sending: {title}, {author}, {year_of_publication}")

    rreply = requests.post('http://localhost:5000/sell/', json={'title': title, 'author': author, 'year_of_publication': year_of_publication})

    print(rreply.status_code)
    print(rreply.json())
```
